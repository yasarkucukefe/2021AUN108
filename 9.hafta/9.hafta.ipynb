```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x**2
```

```python
y
```

```python
# Matplotlib ile grafik
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r')
```

```python
plt.plot(x, y, 'r') # r: red (kırmızı)
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.title('Grafik adı')
```

```python
# Birden fazla grafik çizimi
plt.subplot(1,2,1) # 1: bir satır, 2: iki kolon, 1.grafik çizilecek sonraki komut ile
plt.plot(x, y, 'r')
plt.subplot(1,2,2) # 2. grafik çizilecek sonraki komut ile
plt.plot(y, x, 'g')
```

```python
plt.plot(x, y, 'r--') # r: red (kırmızı), kısa çizgiler ile grafik
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.title('Grafik adı')
```

```python
plt.plot(x, y, 'r*-') # r: red (kırmızı), veri noktalarını * ile göster çizimde
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.title('Grafik adı')
```

```python
plt.plot(x, y, 'r+-') # r: red (kırmızı), veri noktalarını * ile göster çizimde
plt.xlabel('X ekseni')
plt.ylabel('Y ekseni')
plt.title('Grafik adı')
```

# Alternatif grafik çizim yöntemi
## Nesne tabanlı grafik çizimi

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x, y, 'b')
axes.set_xlabel('X ekseni')
axes.set_ylabel('Y ekseni')
axes.set_title('Grafik adı')
```

```python
# Grafik boyutları
fig = plt.figure(figsize=(10,6)) # figsize: grafik boyutlarını belirler

axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x, y, 'b')
axes.set_xlabel('X ekseni')
axes.set_ylabel('Y ekseni')
axes.set_title('Grafik adı')
```

```python
# İç-içe çoklu grafikler
fig = plt.figure(figsize=(10,6))

eksen_1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
eksen_2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

# 1. grafik
eksen_1.plot(x, y, 'r*--')
eksen_1.set_xlabel('1. grafik X ekseni')
eksen_1.set_ylabel('1. grafik Y ekseni')
eksen_1.set_title('1. Grafik başlık')

# 2. grafik (küçük olan)
eksen_2.plot(y, x, 'b')
eksen_2.set_xlabel('2. grafik X')
eksen_2.set_ylabel('2. grafik Y')
eksen_2.set_title('2. grafik başlık')
```

# Çoklu değişken grafikleri

```python
# f1 = x**2-3
# f2 = 2x**3-1

fig = plt.figure(figsize=(8,5));
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2-3, label='f1=x**2-3')
ax.plot(x, 2*x**3-1, label='f2=2x**3-1')
ax.legend(loc=0)# sol üst köşede legend
ax.grid()
ax.set_title("Çoklu Grafik Gösterimi")
```

```python
fig, ax = plt.subplots()
ax.plot(x, x+1, color='blue', linewidth=3)
ax.plot(x, x+2, color='red', linestyle='--', marker='*')
```

# Seaborn grafik kütüphanesi


```python
import seaborn as sns
```

```python
df = pd.read_csv("https://plantdxcdn.s3-us-west-2.amazonaws.com/titanic.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.countplot('Survived',data=df)
```

```python
df.groupby(['Survived'])['Survived'].count()
```

```python
pd.crosstab(df.Pclass,df.Survived)
```

```python
sns.countplot('Pclass',data=df)
```

```python
sns.countplot('Pclass', data=df, hue='Survived')
```

```python
plt.figure(figsize=(10,6))
sns.countplot('Pclass', data=df, hue='Survived')
```

```python
sns.countplot('Sex',data=df, hue='Survived')
```

```python
sns.factorplot('Pclass','Survived',data=df, hue='Sex')
```

```python
sns.factorplot('Sex','Survived',data=df, hue='Pclass')
```

```python
sns.violinplot('Pclass','Age',data=df, hue='Sex')
```

```python
sns.boxplot('Pclass','Age',data=df, hue='Sex')
```

```python
sns.swarmplot('Pclass','Age',data=df, hue='Survived')
```

```python
sns.heatmap(df[['Age','Pclass','Survived']].corr())
```

```python
df['Sex'].unique()
```

```python
df.Sex.replace(('male','female'), (0,1), inplace = True)
```

```python
df['Sex'].unique()
```

```python
sns.heatmap(df[['Survived','Age','Pclass','Sex']].corr(),annot=True)
plt.title('Isı Grafiği')
```
